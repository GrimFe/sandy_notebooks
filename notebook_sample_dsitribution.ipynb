# DISTRIBUTION SAMPLING TEST

```python
import sandy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

- ## Normal distribution

```python
sampled_normal = sandy.sample_distribution(dim=9, nsmp=10000, pdf='normal')
print(f"""mean: {sampled_normal.mean()}
std: {sampled_normal.std()}""")
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15,10))
ax = np.reshape(axs, (1,9))[0]

groups = 30
for i in range(len(ax)):
    count, bins, _ = ax[i].hist(sampled_normal[i], groups, density=True)
    ax[i].plot(bins, 1/(np.std(sampled_normal[i]) * np.sqrt(2 * np.pi)) *
        np.exp( - (bins - np.average(sampled_normal[i]))**2 / (2 * np.std(sampled_normal[i])**2)),
        linewidth=2, color='r')
    ax[i].set_title(f"Distribution {i+1}")
fig.tight_layout()
```

- ## Uniform distribution

```python
sampled_uniform = sandy.sample_distribution(dim=9, nsmp=10000, pdf='uniform')
print(f"""mean: {sampled_uniform.mean()}
std: {sampled_uniform.std()}""")
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15,10))
ax = np.reshape(axs, (1,9))[0]

groups = 30
for i in range(len(ax)):
    count, bins, _ = ax[i].hist(sampled_uniform[i], groups, density=True)
    ax[i].plot(bins, count.mean()*np.ones(groups+1), linewidth=2, color='r')
    ax[i].set_title(f"Distribution {i+1}")
fig.tight_layout()
```
